Splitting the body into codes and words

```python
import re
import string  # Import the string module for English punctuations
from bs4 import BeautifulSoup

def preprocess_data(df, body_column_name, code_column_name, body_words_column_name):
    df[body_words_column_name] = ''
    new_code_column_name = code_column_name + "_all_code"
    df[new_code_column_name] = ''

    for index, row in df.iterrows():
        if not pd.isna(row[body_column_name]):
            print("Value before substitution:", row[body_column_name])
            print("Type before substitution:", type(row[body_column_name]))

            row[body_column_name] = re.sub(r'[\n\t\r\f]', '', row[body_column_name])

            print("Value after substitution:", row[body_column_name])

            soup = BeautifulSoup(row[body_column_name], 'html.parser')
            text_without_code = ''
            all_code_snippets = []

            for pre in soup.find_all('pre'):
                all_code_snippets.extend(code.get_text() for code in pre.find_all('code'))
                pre.decompose()

            for code in soup.find_all('code'):
                all_code_snippets.append(code.get_text())

            text_without_code = soup.get_text()
            text_without_code = text_without_code.translate(str.maketrans('', '', string.punctuation))

            df.at[index, body_words_column_name] = text_without_code  # Assign to body_words_column_name using index
            df.at[index, new_code_column_name] = all_code_snippets  # Assign to new_code_column_name using index

    return df
```

Cleaning the "body_words" column with nkltTokeniser

```python
# Function definition:

def preprocessing_task2(text):

    text = text.lower()

    words = word_tokenize(text)

    stop_words = set(stopwords.words('english'))

    filtered_words = [word for word in words if word not in stop_words]

    stemmed_words = [stemmer.stem(word) for word in words]

    stemmed_text = ' '.join(stemmed_words)

    return stemmed_text
```
